```python
#Extract the required feature into feature variable
features=['genres','keywords','director','cast']

#remove null values of homepage and tagline

feat=['homepage','tagline','overview','release_date','runtime']

for feats in feat:
    df[feats]=df[feat].fillna('')
    
```

```python
for feature in features:
    df[feature]=df[features].fillna('')
```

```python
df.isnull().sum()
```

```python
#Define function to combine features into single comlumn
def combine_feature(row):
    return str(row['genres'])+""+str(row['keywords'])+""+str(row['director'])+""+str(row['cast'])

df['combined features']=df.apply(combine_feature,axis=1)
```

#Data vectorization

```python
#Tokenization and vectorization

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
# Prints dict of key-value pair
print(cv.vocabulary_)
```

```python
# word counting
x=cv.fit_transform(df['combined features']) #x is a sparse matrix
print(x.toarray())
print(type(x))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
```

```python
A_sparse=sparse.csr_matrix(x) #convert sparse matrix to cosine Similarity Matrix
similarity_score=cosine_similarity(A_sparse,dense_output=True) #if False then sparse output is returned
```

```python
print(similarity_score)
```

```python
df['index']
```

```python
def gt_title(tit):
    return df[df.index==tit]['title'].values[0]

def gt_index(ind):
    return df[df.title==ind]['index'].values[0]
```

```python
#take value from user and search for similiar movies
input_user=input('\n Enter a movie name \n')
movie_index=gt_index(input_user)
similiar_movies=list(enumerate(similarity_score[movie_index]))
```

```python
sorted_similiar_movies=sorted(similiar_movies,key=lambda x:x[1],reverse=True)[1:] #Descending values
```

```python
i=0
print('Top 5 similiar movies are \n')
for element in sorted_similiar_movies:
    print(gt_title(element[0]))
    
    i=i+1
    if i>5:
        break
```
